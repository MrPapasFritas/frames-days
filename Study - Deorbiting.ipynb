# Study - Deorbiting
This code intends to study the deorbiting strategies for a 3U Cubesat through orbital mechanics.  
The variables of interest are:
<ul>
    <li><b>Propellant Usage:</b> the approximate amount of propellant used for the different deorbiting strategies.</li>
    <li><b>Energy:</b> the approximate amount of power and energy used by the thruster in order to perform the maneuver.</li>
    <li><b>Time:</b> the amount of time needed for deorbiting on each scenario.</li>
</ul>

```python
from maneuvers import Maneuvers
import auxiliary
import constants
import models
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime
%load_ext jupyternotify
%load_ext autoreload
%autoreload 2
```

## Case 1: Free Fall
This scenario evaluates how much time it takes to fall from LEO

```python
# INITIAL CONDITIONS
coe,date = auxiliary.parseTle("suchai0.tle")
T = 2*np.pi*(coe[0]**3/constants.mu_E)**0.5
print("Period: ",T/60," min")
```

### 1.1 Free Fall with all perturbations

```python
# SPACECRAFT DEFINITIONS
cubesat = models.Cubesat("3U")

# MANEUVERING
naturalFall = Maneuvers(coe,cubesat,date,verbose=True)
# A Month of perturbations
naturalFall.addPerturbation("J2")
naturalFall.addPerturbation("atmosphere")
naturalFall.addPerturbation("solar_pressure")
naturalFall.addPerturbation("moon_gravity")
naturalFall.addPerturbation("sun_gravity")

naturalFall.propagate(60*60*24*30*12*3.75,60*10)
%notify
```

```python
%config InlineBackend.figure_format = 'retina'
naturalFall.makeReport();
naturalFall.plot("coe");
naturalFall.plot("singleItem",np.linalg.norm(naturalFall.history.r,axis=1)/1e3-constants.Re/1e3);
plt.ylabel("Altitude [km]");
```

## Case 2: Freefall + Thrusting to shrink semi-major axis (a)

```python
# SPACECRAFT DEFINITIONS
cubesat = models.Cubesat("3U")
cubesat.thruster = models.NanoPropCGP3()
cubesat.dryMass = cubesat.wetMass-500e-3

# MANEUVERING
thrustedFall = Maneuvers(coe,cubesat,date,verbose=True)
thrustedFall.addPerturbation("atmosphere")
thrustedFall.addPerturbation("thrust")
def alphaCallback(coe):
    e = coe[1]
    nu = coe[5]
    return np.arctan2(e*np.sin(nu),1+e*np.cos(nu))+np.pi
def betaCallback(coe):
    return 0
thrustedFall.thrustProfile = (alphaCallback,betaCallback)
thrustedFall.propagate(60*60*24*5,60)
%notify
```

```python
# RESULTS
%config InlineBackend.figure_format = 'retina'
thrustedFall.makeReport();
thrustedFall.plot("coe");
thrustedFall.plot("singleItem",np.linalg.norm(thrustedFall.history.r,axis=1)/1e3-constants.Re/1e3);
plt.ylabel("Altitude [km]");
thrustedFall.plot("singleItem",thrustedFall.history.propMass)
plt.title("Propellant Usage")
plt.ylabel("Remaining Propellant Mass [kg]");
```

## Case 3: Freefall + Thrusting to increment eccentricity (e)

```python
# SPACECRAFT DEFINITIONS
cubesat = models.Cubesat("3U")
cubesat.thruster = models.NanoPropCGP3()
cubesat.dryMass = cubesat.wetMass-700e-3

# MANEUVERING
thrustedFall_e = Maneuvers(coe,cubesat,date,verbose=True)
thrustedFall_e.addPerturbation("atmosphere")
thrustedFall_e.addPerturbation("thrust")
def alphaCallback(coe):
    e = coe[1]
    nu = coe[5]
    E = 2*np.arctan2(np.tan(nu/2),np.sqrt((1+e)/(1-e)))
    return np.arctan2(np.sin(nu),np.cos(nu)+np.cos(E))
def betaCallback(coe):
    return 0
thrustedFall_e.thrustProfile = (alphaCallback,betaCallback)
thrustedFall_e.propagate(60*60*24*10.2,60)
%notify
```

```python
# RESULTS
%config InlineBackend.figure_format = 'retina'
thrustedFall_e.makeReport();
thrustedFall_e.plot("coe");
thrustedFall_e.plot("singleItem",np.linalg.norm(thrustedFall_e.history.r,axis=1)/1e3-constants.Re/1e3);
plt.ylabel("Altitude [km]");
thrustedFall_e.plot("singleItem",thrustedFall_e.history.propMass)
plt.title("Propellant Usage")
plt.ylabel("Remaining Propellant Mass [kg]");
```
